## Lab - 선형 회귀를 위한 경사하강법 구현 (P22)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0.0, 1.0, 2.0])
y = np.array([3.0, 3.5, 5.5])
plt.plot(X, y, "r.")
plt.grid(True)
plt.ylim([0,6])
```

```python
W = []
b = []
W.append(0)
b.append(0)
lrate = 0.01
epochs = 100
n = len(X)
```

```python
for t in range(epochs):
    y_pred = W[-1]*X + b[-1]
    diff = y_pred - y
    dW = 2/n * np.sum(diff * X)
    db = 2/n * np.sum(diff)
    W.append(W[-1] - lrate * dW)
    b.append(b[-1] - lrate * db)
```

```python
print("W:", W)
print("b:", b)
```

```python
for t in range(epochs):
    y_pred = W[t]*X + b[t]
    plt.plot(X, y_pred)
plt.plot(X, y, 'r.')
plt.grid(True)
plt.ylim([0,6])
```

## BMI와 혈당치에 대한 선형 회귀 문제 -> 사이킷런 라이브러리 활용 (P25~27)

### 데이터 준비

```python
from sklearn import datasets
```

```python
# 당뇨병 데이터 세트를 적재
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
# 하나의 특징(BMI)만 추려내서 2차원 배열로 제작 (cf. BMI 특징의 인덱스 : 2)
diabetes_X_new = diabetes_X[:, np.newaxis, 2]
```

```python
# 학습 데이터와 테스트 데이터를 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes_X_new, diabetes_y, test_size=0.1, random_state=0)
```

### 사이킷런 라이브러리를 통한 선형회귀 구현

```python
import matplotlib.pylab as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color = 'black')
plt.plot(X_test, y_pred, color = 'red', linewidth = 3)
plt.show()
```
