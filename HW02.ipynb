## BMI와 혈당치에 대한 선형 회귀 문제 -> 사이킷런 라이브러리 활용

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 키와 BMI에 따른 몸무게 데이터
X = np.array([1.50, 1.60, 1.70, 1.80, 1.90]) # 키(m)
y = np.array([50, 60, 70, 80, 90]) # 몸무게(kg)
```

```python
# 그래프 초기 설정
plt.plot(X, y, 'r.')
plt.grid(True)
plt.ylim(40, 100)
```

```python
# 가중치(W), 바이어스(b) 초기화
W = []
b = []
W.append(0) # 초기 가중치(W)
b.append(0) # 초기 바이어스(b)
```

```python
# 학습룰, 반복 횟수 설정
lrate = 0.01 # 학습률
epochs = 100 # 반복 횟수
n = len(X) # 데이터 포인터 수
```

```python
# 경사 하강법으로 가중치와 절편을 업데이트
for i in range(epochs):
    y_pred = W[-1] * X + b[-1] # 예측 값
    diff = y_pred - y # 실제 값과의 차이

    dW = 2/n * np.sum(diff * X) # W에 대한 편미분
    db = 2/n * np.sum(diff) # b에 대한 편미분

    W.append(W[-1] - lrate * dW) # W 업데이트
    b.append(b[-1] - lrate * db) # b 업데이트
```

```python
print('W: ', W)
print('b: ', b)
```

```python
# 예측 결과 시각화
for t in range(epochs):
    y_pred = W[t] * X + b[t]
    plt.plot(X, y_pred)

plt.plot(X, y, 'r.')
plt.grid(True)
plt.ylim([40, 100])
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.show
```
